```python
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib

matplotlib.rcParams['font.family'] = "serif"
%config InlineBackend.figure_format = 'retina'
```

```python
final_test_scores = np.array([[0.7615, 0.7729, .7642],
                             [0.8520, 0.8964, 0.9029],
                             [0.7547, 0.7586, 0.7628],
                             [0.8453, 0.8920, 0.9007],
                             [0.8735, 0.9091, 0.9164],
                             [0.8830, 0.8523, 0.8038],
                             [0.7536, 0.7528, 0.7495]])
labels = np.array(['ADA', 'GBT', 'L-SVM', 'R-SVM', 'RF', 'KNN', 'LogReg'])

order = np.argsort(np.mean(final_test_scores, axis=1))[::-1]
```

```python
plt.subplots(figsize=(3.75,3))
g = sns.heatmap(final_test_scores[order], cmap='RdBu', annot=True, 
                fmt='.4f', xticklabels=[0.2, 0.5, 0.8],
               linewidths=0.5, cbar=False)
g.set_yticklabels(labels[order], rotation=0, fontname="serif");
plt.subplots_adjust(left=0.2, bottom=0.2, right=0.98, top=0.98, wspace=0, hspace=0);
plt.xlabel('Data Partition');
plt.savefig('average_score.png', dpi=300);
```
